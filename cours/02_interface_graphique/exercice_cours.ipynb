```python
import tkinter as tk

racine = tk.Tk()
racine.title("tk")

gauche = tk.Label(racine, text="Choisir une valeur pour l'opérande de gauche", font=("helvetica", "26"))
droite = tk.Label(racine, text="Choisir une valeur pour l'opérande de droite", font=("helvetica", "26"))
calculer = tk.Label(racine, text="calculer", font=("helvetica", "26"))
elt_gauche = tk.Label(racine, text="12", font=("helvetica", "26"))
addition = tk.Label(racine, text="+", font=("helvetica", "26"))
elt_droite = tk.Label(racine, text="25", font=("helvetica", "26"))
egal = tk.Label(racine, text="=", font=("helvetica", "26"))
reponse = tk.Label(racine, text="37", font=("helvetica", "26"))

gauche.grid(row= 1, column= 0)
droite.grid(row= 2, column= 0)
calculer.grid(row= 1, column= 1, columnspan= 5)

elt_gauche.grid(row= 0, column= 1)
addition.grid(row= 0, column= 2)
elt_droite.grid(row= 0, column= 3)
egal.grid(row= 0, column= 4)
reponse.grid(row= 0, column= 5)

racine.mainloop()
```

```python
import tkinter as tk

def op_gauche():
    global num_gauche
    num_gauche = int(input(""))
    elt_gauche.config(text= num_gauche)

def op_droite():
    global num_droite
    num_droite = int(input(""))
    elt_droite.config(text= num_droite)

def calc():
    add = num_gauche + num_droite
    reponse.config(text= add)

racine = tk.Tk()
racine.title("tk")

gauche = tk.Button(racine, text="Choisir une valeur pour l'opérande de gauche",
                   command= op_gauche, font=("helvetica", "26"))
droite = tk.Button(racine, text="Choisir une valeur pour l'opérande de droite",
                   command= op_droite, font=("helvetica", "26"))
calculer = tk.Button(racine, text="calculer",
                     command= calc, font=("helvetica", "26"))

elt_gauche = tk.Label(racine, text="12", font=("helvetica", "26"))
addition = tk.Label(racine, text="+", font=("helvetica", "26"))
elt_droite = tk.Label(racine, text="25", font=("helvetica", "26"))
egal = tk.Label(racine, text="=", font=("helvetica", "26"))
reponse = tk.Label(racine, text="37", font=("helvetica", "26"))

gauche.grid(row= 1, column= 0)
droite.grid(row= 2, column= 0)
calculer.grid(row= 1, column= 1, columnspan= 5)
elt_gauche.grid(row= 0, column= 1)
addition.grid(row= 0, column= 2)
elt_droite.grid(row= 0, column= 3)
egal.grid(row= 0, column= 4)
reponse.grid(row= 0, column= 5)

racine.mainloop()
```
